Here in this intent i created different books of different genre, some of them have prices and some of them do not have prices

* A checkout & fine section is also provided in this task

```python
intents  = {
    "greetings": {
        "patterns":["hi","hello","halo","hey"],
        "responses":["hey there, how may i help you?","hi, nice to meet you, how can i help you?","hello, how can I help you?"]
    },
    "books": {
        "patterns": ["book","availablebooks","books"],
        "responses": ["what kind of books?", "genre ?"]
    },
    "genres": {
        "patterns": ["genre","genres"],
        "responses": ["we have: Adventure, fairy tales, scifi, english vocabulary, maths, news articles, comics ,etc."]
    },
    "adventure": {
        "patterns": ["adventure","adventurebooks","another_adventure"],
        "books": {"percy jackson(All volumes)":850,
                      "Harry Potter (All volumes)":615,
                      "Alchemist": 392,
                      "Madagascar Life": 256}
    },
    "fairy tales": {
        "patterns": ["tales","fairy","fairytales"],
        "books": {"SnowWhite":134,
                      "Malfecient": 159,
                      "Cindrella": 81}
    },
    "scifi": {
        "patterns": ["scifi","sci-fi","sci_fi"],
        "books": {"Terminator 2204": 420,
                      "Resident-Evil":400,
                      "Darth Vader": 390}
    },
    "english": {
        "patterns": ["english","englishvocabulary","vocab","vocabulary"],
        "books": {"vocabulary skills":0,
                      "presentation skills":6 ,
                      "listening skills": 0,
                      "reading skills": 0}
    },
    "maths": {
        "patterns": ["maths","math"],
        "books": {"trignometric rules": 0,
                      "Differentiation & Integration Vol 1,2,3":0,
                      "Basic Maths for everyone": 0,
                      "Probability in Machine Learning": 0,
                      "Mathematical Algorithms": 0}
    },
    "news articles": {
        "patterns": ["news","newspaper","newsarticles", "newseditorial","editorials"],
        "books": {"The Hindu":0,
                      "Times of India": 0,
                      "HinduEditorials":0,
                      "Digest":0}
    },
    "comics": {
        "patterns": ["comic","comics"],
        "books": {"Daredevil Reborn 2": 164,
                      "Batman Legacy": 176,
                      "Young Megamind": 200,
                      "Spiderman: The Wanted": 280}
    },
    "time": {
        "patterns" : ["time","closingtime"],
        "responses": ["Library is open from 9AM to 6PM"]
    },
    "checkout":{
        "patterns": ["checkout","check_out","check-out"],
        "responses": []
    }
}
```

```python
import random

cart = []

def get_intent(user_input):
    for intent, data in intents.items():
        for pattern in data["patterns"]:
            if pattern in user_input:
                return intent
    return "Unknown"


def chatbot():
    print("Chatbot: Welcome, I am Student and College Library Assistant.")
    print("Type exit to end the chat\n")

    while True:
        try:
            user_input = input("User: ").lower()

            if user_input == "exit":
                print("Chatbot: Goodbye! Have a great day.")
                break

            intent = get_intent(user_input)

            # Fine calculation
            if "fine" in user_input:
                print("Chatbot: Enter the number of late days:")
                days = int(input("User: "))
                total_fine = days * 2
                print(f"Chatbot: Your total fine is ₹{total_fine}.")

            # Book categories
            elif intent in ["adventure", "fairy tales", "scifi", "english", "maths", "news articles", "comics"]:
                books = intents[intent]["books"]

                print("Chatbot: Available books:")
                book_list = list(books.items())

                for i, (book, price) in enumerate(book_list, start=1):
                    print(f"{i}. {book} - ₹{price}")

                print("Enter book number to add to cart (or 0 to skip):")

                try:
                    choice = int(input("User: "))

                    if choice == 0:
                        continue

                    if 1 <= choice <= len(book_list):
                        selected_book, price = book_list[choice - 1]
                        cart.append((selected_book, price))
                        print(f"Chatbot: '{selected_book}' added to cart.")
                    else:
                        print("Chatbot: Invalid choice.")

                except ValueError:
                    print("Chatbot: Please enter a valid number.")

            # Checkout
            elif intent == "checkout":
                if not cart:
                    print("Chatbot: Your cart is empty.")
                else:
                    total = sum(price for _, price in cart)

                    print("Chatbot: Your cart:")
                    for book, price in cart:
                        print(f"- {book}: ₹{price}")

                    print(f"Total Price: ₹{total}")

            # Unknown intent
            elif intent == "Unknown":
                print("Chatbot: I'm sorry, I didn't understand that. Can you please rephrase?")

            # Normal responses (ONLY if responses exist)
            elif "responses" in intents[intent]:
                response = random.choice(intents[intent]["responses"])
                print("Chatbot:", response)

        except ValueError:
            print("Chatbot: Please enter a valid number (e.g., 1, 2, 3).")

        except KeyError:
            print("Chatbot: Something went wrong with intent mapping.")

        except Exception as e:
            print("Chatbot: An unexpected error occurred:", str(e))

        finally:
            print("Chatbot: Ready for next input...\n")


chatbot()
```
