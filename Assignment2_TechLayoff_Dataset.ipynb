# Performing **Data Manipulation**

```python
layoff_region=data[data["region"]== "Europe"]   # layoff occured in Europe region
print(layoff_region)
```

```python
jobscut= data[data["jobs_cut"]>2000]
print(jobscut)
```

```python
# jobs_cut with ai_cited only and the jobs_cut greater than 2000
aicited_jobscut=data[(data["jobs_cut"]>2000) & (data["ai_cited"]==True)]
print(aicited_jobscut)
```

```python
layoffs_verifiedsource=data[data["verified_source"].isin(['CNBC','IBTimes','CNBC / Crunchbase'])]
print(layoffs_verifiedsource)
```
