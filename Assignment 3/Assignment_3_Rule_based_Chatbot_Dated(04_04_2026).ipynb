Assignment: Our assignment is to add commodity, commodityprice data as an Intent and our Chatbot must give us a reply for a user query.

```python
intents= {
    "greeting": {
        "patterns": ["hi", "hey", "hello", "good morning", "good evening"],
        "responses": ["Hello there!", "Hi!", "Hey!", "Greetings!"]
    },
    "price_query": {
        "patterns": ["price", "cost", "value", "rate"],
        "responses": ["Please tell me the commodity name"]
    },
    "goodbye": {
        "patterns": ["bye", "goodbye", "see you later"],
        "responses": ["Goodbye!", "see you later"]
    }
    }


commodities = {
    "potato": {
        "patterns": ["potato", "potatoes", "pot"],
        "responses": ["yes we have","sure","ofcourse","it is available"]
    },
    "tomato": {
        "patterns": ["tomato", "tomatos", "toma"],
        "responses": ["yes we have", "sure","ofcourse","available"]
    },
    "onion": {
        "patterns": ["onion", "onions", "oni"],
        "responses": ["yes we have","sure","ofcourse", "yes, it is available"]
    },
    "carrot": {
        "patterns": ["carrot", "carrots", "caro"],
        "responses": ["yes we have","sure","ofcourse"]
    }
}

commodity_prices = {
    "potato_price": {
        "patterns": ["potatoprice", "potato_price","priceofpotato"],
        "responses": [" It cost ₹20 per kg", " ₹20/kg"]
    },
    "tomato_price": {
        "patterns": ["tomatoprice", "tomato_price","priceoftomato"],
        "responses": [" It cost ₹30 per kg", " ₹18/kg"]
    },
    "onion_price": {
        "patterns": ["onionprice", "onionsprice", "priceofonion"],
        "responses": ["It cost ₹20 per kg", " ₹20/kg"]
    },
    "carrot_price": {
        "patterns": ["carrotprice", "carrot_price", "priceofcarrot"],
        "responses": ["It cost ₹15 per kg", " ₹15/kg"]
    }
}
```

```python
# intent detection function
import random
def get_intent(user_input):
  user_input = user_input.lower()
  for intent,data in intents.items():
    for pattern in data["patterns"]:
      if pattern in user_input:
        return intent
  return "Unknown"

def get_commodity(user):
  user = user.lower()
  for commodity,data in commodities.items():
    for pattern in data["patterns"]:
      if pattern in user:
        return commodity


def get_commodity_price(user_input):
  user_input = user_input.lower()
  for commodity_price,data in commodity_prices.items():
    for pattern in data["patterns"]:
      if pattern in user_input:
        return commodity_price



def chatbot():
  print("Chatbot: Hello! How can I assist you today?")
  while True:
    user_input = input("You: ")
    intent=get_intent(user_input)
    commodity=get_commodity(user_input)
    commodity_price=get_commodity_price(user_input)


    if commodity_price:
      response=random.choice(commodity_prices[commodity_price]["responses"])
      print("Chatbot:",response)
      continue

    if commodity:
      response=random.choice(commodities[commodity]["responses"])
      print("Chatbot:",response)
      continue

    if intent=="Unknown":
      print("Chatbot: I'm sorry, I didn't understand that. Can you please rephrase?")
      continue

    if intent=="goodbye":
      print("Chatbot: Goodbye! Have a great day!")
      break


    else:
      responds = random.choice(intents[intent]["responses"])
      print("Chatbot:",responds)



chatbot()
```
