# Performing **Data Manipulation**

```python
layoff_region=data[data["region"]== "Europe"]   # layoff occured in Europe region
print(layoff_region)
```

```python
jobscut= data[data["jobs_cut"]>2000]
print(jobscut)
```

```python
# jobs_cut with ai_cited only and the jobs_cut greater than 2000
aicited_jobscut=data[(data["jobs_cut"]>2000) & (data["ai_cited"]==True)]
print(aicited_jobscut)
```

```python
layoffs_verifiedsource=data[data["verified_source"].isin(['CNBC','IBTimes','CNBC / Crunchbase'])]
print(layoffs_verifiedsource)
```

# Data Visualization

```python
plt.plot(data["layoffs_2025"])
plt.title("Layoffs in 2025")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.scatter(data["layoffs_2024"],data["layoffs_2025"])
plt.title("Layoffs in 2024 vs Layoffs in 2025")
plt.xlabel("2024 Layoffs")
plt.ylabel("2025 Layoffs")
plt.show()
```

```python
plt.hist(data["layoff_size_category"], bins=30)
plt.title("Layoff Size Category")
plt.xlabel("Size Category")
plt.ylabel("Frequency")
plt.show()
```

```python
data["sector"].value_counts().plot(kind="bar")
plt.title("bar chart of sector layoffs")
plt.xlabel("sector")
plt.ylabel("frequency or count")
plt.show()
```

```python
plt.boxplot(data['company_revenue_2025_bn'])
plt.title("boxplot of company revenue in 2025")
plt.xlabel("company revenue")
plt.show()
```

```python
# pie chart
data["hq_city"].value_counts().plot(kind="pie", autopct="%0.1f%%")
plt.title("pie chart of layoffs occured in city headquarters")
plt.show()
```

```python
sns.histplot(x="pct_workforce_cut",data=data)
plt.show()
```

```python
sns.heatmap(data.iloc[:,9:12].corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap")
plt.show()
```
